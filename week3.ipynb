```python
import cv2 as cv
from google.colab.patches import cv2_imshow #for image display
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
print(' Mengubah tingkat kecerahan citra')
print(' -------------------------------- ')
try :
    brightness = int(input('Masukan nilai kecerahan : '))
except ValueError:
    print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

#akses per pixel

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

#cara simple tanpa for loop

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```python
print(' Mengubah kontras dan tingkat kecerahan citra')
print('---------------------------------------------')
try:
  brightness = int(input('Masukkan tingkat kecerahan[-255 - 255]: '))
  contrast = float(input('Masukkan kontras [1.0-3.0]: '))
except ValueError:
  print('Error, not a number')

image = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')
kontras_image = np.zeros(image.shape, image.dtype)

for y in range(kontras_image.shape[0]):
  for x in range(kontras_image.shape[1]):
    for c in range(kontras_image.shape[2]):
      kontras_image[y,x,c] = np.clip(contrast*image[y,x,c] + brightness,0, 255)

final_frame = cv.hconcat((image, kontras_image))
cv2_imshow(final_frame)
```

```python
averaging_image = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')
asli = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')

averaging_image[:] = averaging_image.mean(axis=-1, keepdims=1)
final_frame2 = cv.hconcat((asli, averaging_image))
cv2_imshow(final_frame2)
print('hasil dengan plotlib')
plt.imshow(averaging_image)
```

```python
light_image = cv.imread('/content/drive/MyDrive/kucingkuu.jpg')

light_image[:] = np.max(light_image,axis=-1,keepdims=1)/2+np.min(light_image,axis=-1,keepdims=1)/2
final_frame3 = cv.hconcat((asli, light_image))
cv2_imshow(final_frame3)
print('hasil dengan plotlib')
plt.imshow(light_image)
```
